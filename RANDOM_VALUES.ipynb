## Introduce and handle missing values



```python

np.random.seed(42)
rows_to_nan = np.random.choice(ship.index, size=100, replace=False)
ship.loc[rows_to_nan, 'age'] = np.nan
print("Missing values before filling:")
print(ship.isnull().sum())

mean_age = ship['age'].mean()
print(f"\nMean age: {mean_age}")

ship['age'].fillna(mean_age, inplace=True)

print("\nMissing values after filling:")
print(ship.isnull().sum())
```
